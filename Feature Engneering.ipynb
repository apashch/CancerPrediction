Here I did some manual looking up in the NHANES variable descriptions and constructed a roadmap in 'SVM_selected' table

```python
def process_numerical(col):
    colv = col[~col.isna()].values
    # replace really small numbers with 0s
    colv[colv < 10e-10] = 0
    
    #normalize
    colv = (colv-colv.mean())/colv.std()
    
    col[~col.isna()] = colv
    
    print(col.tail())
    return col
    
def process_binary(col, truevals = [1]):
    colv = col[~col.isna()].values
    
    truemask = np.isin(colv, truevals)
    colv[~truemask] = 0
    colv[truemask]  = 1
    
    col[~col.isna()] = colv
    return col

def process_categorical(col):
    return pd.get_dummies(col, prefix=col.name, prefix_sep='#')
```

```python
dropcols = "FIAPROXY PHQ050 PHQ060 BPQ050A"
dropcols = dropcols.split(" ")
selected.drop(dropcols, axis= 1,inplace = True)

numericalcols = "DMDHRAGE RIDAGEYR DR1TKCAL DR1TM161 DR1TS100 DR1TS140 DR1TSUGR DR2TNIAC DR2TPROT DR1_320Z DR1_330Z DR1BWATZ BPXSY2 BPXSY1 BPXDI2 BPXDI1 BMXARMC BMXARML BMXHT LBXHBS LBDLYMNO LBDMONO LBDNENO LBXMCHSI LBXMPSI LBDSCASI LBDSTPSI LBXSCA LBXSNASI LBXSTP LBXRDW HSD010"
numericalcols = numericalcols.split(" ")
for colname in numericalcols:
    selected.loc[:,colname] =  process_numerical(selected[colname])
    
binarycols = "DMQMILIT BPQ040A KIQ022 MCQ010 MCQ160N PAQ620 PFQ059 HUQ090 BPAEN2 BPXPTY SXQ260 SXQ265 SXQ272"
binarycols = binarycols.split(" ")
for colname in binarycols: 
    selected.loc[:,colname] =  process_binary(selected[colname])
    
binaryspecial = "DR1HELPD DR2HELPD"
binaryspecial = binaryspecial.split(" ")
for colname in binaryspecial:
    selected.loc[:,colname] =  process_binary(selected[colname], [1, 13])
```

```python
selected["DR1_WATER"] = selected["DR1_320Z"]+selected["DR1_330Z"]+selected["DR1BWATZ"]
selected.drop(columns = ["DR1_320Z","DR1_330Z","DR1BWATZ"], inplace = True)

selected["SXQ"] = selected["SXQ260"]+selected["SXQ265"]+selected["SXQ272"]
selected["SXQ"] = selected["SXQ"]>=1
selected.drop(columns = ["SXQ260","SXQ265", "SXQ272"], inplace = True)
```

### Let's add some general demographic variables that may influence the results

```python
demo=all_features[["RIAGENDR", "RIDAGEYR", "RIDRETH1", "DMDEDUC2", "BMXBMI", "BMXWT", "BMXHT"]]
```

```python
# drop  what's already in the dataset to avoid repeating
dropcols = "RIDAGEYR"
dropcols = dropcols.split(" ")
demo.drop(dropcols, axis= 1,inplace = True)

numericalcols = "DMDEDUC2 BMXBMI BMXWT BMXHT"
numericalcols = numericalcols.split(" ")
for colname in numericalcols:
    demo.loc[:,colname] =  process_numerical(demo[colname])
    
binarycols = "RIAGENDR"
binarycols = binarycols.split(" ")
for colname in binarycols:
    demo.loc[:,colname] =  process_binary(demo[colname])
    
catcols = "RIDRETH1"
catcols = catcols.split(" ")
for colname in catcols:
    demo = pd.concat([demo, process_categorical(demo[colname])], axis = 1)
demo.drop(columns = catcols, inplace = True)
```

```python
demo_nona = demo.fillna(demo.median())
```
