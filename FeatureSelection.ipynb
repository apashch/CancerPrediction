```python
import pandas as pd
import nhanes as nhanes

DATA_PATH = '/Users/Artem/Documents/CS 205/NHANES/'
```

```python
features = pd.read_csv("/Users/Artem/Documents/GitHub/CancerPrediction/all_features.csv")
target = features[features.feature == 'MCQ220']
# drop target
features = features[features.feature != 'MCQ220']
# drop all closely related cols
features = features[~features.feature.str.contains("MCQ22|MCQ23|MCQ24", regex = True)]
```

Drop some other irrelevant stuff

```python
# questions not aksed in adult population
features = features[~features.feature.str.contains("PAQ72|PAQ73|PAQ74|PAQ75|PAQ76|PAQ77", regex = True)]
features = features[~features.file.str.contains("SSPFAC|WHQMEC", regex = True)]

features.head()
```

```python
# select 100 random features from remaining
subset = features.sample(10)
subset = pd.concat([subset, target])
subset.head()
```

```python
ds = nhanes.Dataset(DATA_PATH)

ds.load_custom(subset, verbose = True)
# n_fe = ds.features.shape[1]
# n_classes = 2

# dataset_features = ds.features
# dataset_targets = ds.targets
```
